```python
#import necessary packages

#our workhorses
import numpy as np
import pandas as pd
import scipy

#to visualize
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as patches
#style params for figures
sns.set(font_scale = 2)
plt.style.use('seaborn-white')
plt.rc("axes", labelweight="bold")
from IPython.display import display, HTML

#to load files
import os
import h5py

#for stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import statsmodels.stats.multitest as multitest
import statsmodels.stats.multicomp


#import cusotm functions
from EMG_gestures.utils import *
```

```python
# Load RNN within-subject data
nsubjects = 36

rnn_results_df = [] 

results_folder = '../results_data/single_subject_training/RNN'
for src_subject_id in range(1,nsubjects+1):
    for n_train in np.array([10,20]):
        if n_train == 10:
            results_fn = 'subject_%02d_within_subject_results.h5'%(src_subject_id)
        else:
            results_fn = 'subject_%02d_within_subject_%d_train_sets_results.h5'%(src_subject_id,n_train)
        subject_df = pd.read_hdf(os.path.join(results_folder,results_fn), 'results_df')
        subject_df['n_train'] = n_train
        subject_df['Train_Subject'] = subject_df['Subject']
        subject_df['Test_Subject'] = subject_df['Subject']
        subject_df.drop(columns = ['Subject'])
        rnn_results_df.append(subject_df)

rnn_results_df = pd.concat(rnn_results_df,axis = 0)
```

```python

# Average over multiple train/test splits of the data
rnn_results_df = rnn_results_df.groupby(['Shuffled','Type','Train_Subject','Test_Subject','n_train'],as_index = False)\
.mean()\
.drop(columns = ['Fold','Rep'])

# Output summary
display(HTML(rnn_results_df.groupby(['Shuffled','Type','n_train']).mean().drop(columns = ['Train_Subject','Test_Subject']).to_html()))
```

```python
# Load RNN across-subject data
nsubjects = 36

rnn_xsubj_results_df = [] 

results_folder = '../results_data/single_subject_training/RNN'
for src_subject_id in range(1,nsubjects+1):
    for n_train in np.array([10,20,40]):
        if n_train == 10:
            results_fn = 'subject_%02d_across_subject_results.h5'%(src_subject_id)
        else:
            results_fn = 'subject_%02d_across_subject_%d_train_sets_results.h5'%(src_subject_id,n_train)
        subject_df = pd.read_hdf(os.path.join(results_folder,results_fn), 'results_df')
        subject_df['n_train'] = n_train
        rnn_xsubj_results_df.append(subject_df)

rnn_xsubj_df = pd.concat(rnn_xsubj_results_df,axis = 0)

# # Average over test subjects
rnn_xsubj_df = rnn_xsubj_df.groupby(['n_train','Shuffled','Type','Train_Subject'],as_index = False)\
.mean()\
.drop(columns = ['Test_Subject','Rep'])

# # Output summary
display(HTML(rnn_xsubj_df.groupby(['Shuffled','Type','n_train']).mean().drop(columns = ['Train_Subject']).to_html()))

```
